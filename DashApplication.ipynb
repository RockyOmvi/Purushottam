```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data: Replace this with your actual data
data = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020, 2021],
    'Sales': [2000, 2500, 3000, 3500, 4000, 3000, 4500]
}

df = pd.DataFrame(data)

# Line Chart
plt.figure(figsize=(10, 5))
plt.plot(df['Year'], df['Sales'], marker='o')
plt.title('Automobile Sales Over the Years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid()
plt.show()
```

```python
# Sample data for vehicle types
data = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020, 2021],
    'Sedan': [1000, 1200, 1500, 1800, 2000, 1500, 2500],
    'SUV': [800, 1000, 1200, 1400, 1600, 1200, 2000],
    'Truck': [200, 300, 400, 500, 600, 300, 500],
}

df = pd.DataFrame(data)

# Line Chart for Vehicle Types
plt.figure(figsize=(10, 5))
for vehicle_type in ['Sedan', 'SUV', 'Truck']:
    plt.plot(df['Year'], df[vehicle_type], marker='o', label=vehicle_type)
plt.title('Automobile Sales by Vehicle Type')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend()
plt.grid()
plt.show()
```

```python
import seaborn as sns

# Sample data for recession vs non-recession
data = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020, 2021],
    'Vehicle Type': ['Sedan', 'SUV', 'Truck', 'Sedan', 'SUV', 'Truck', 'Sedan'],
    'Sales': [1000, 800, 200, 2000, 1600, 600, 2500]
}

df = pd.DataFrame(data)

# Seaborn Line Plot
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='Year', y='Sales', hue='Vehicle Type')
plt.title('Sales Trend Comparison: Recession vs Non-Recession')
plt.show()
```

```python
# Sample GDP data
gdp_data = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020, 2021],
    'GDP Recession': [18000, 17500, 18000, 18500, 19000, 17000, 20000],
    'GDP Non-Recession': [19000, 19500, 20000, 20500, 21000, 20500, 22000]
}

gdp_df = pd.DataFrame(gdp_data)

# Subplots
fig, ax = plt.subplots(2, 1, figsize=(10, 10))

ax[0].plot(gdp_df['Year'], gdp_df['GDP Recession'], marker='o', color='red')
ax[0].set_title('GDP During Recession')
ax[0].set_ylabel('GDP')

ax[1].plot(gdp_df['Year'], gdp_df['GDP Non-Recession'], marker='o', color='green')
ax[1].set_title('GDP During Non-Recession')
ax[1].set_ylabel('GDP')

plt.xlabel('Year')
plt.tight_layout()
plt.show()
```

```python
# Sample data for seasonality
season_data = {
    'Season': ['Winter', 'Spring', 'Summer', 'Fall'],
    'Sales': [1000, 2000, 3000, 2500],
    'Vehicle Type': ['Sedan', 'SUV', 'Truck', 'Sedan']
}

season_df = pd.DataFrame(season_data)

# Bubble Plot
plt.figure(figsize=(10, 5))
plt.scatter(season_df['Season'], season_df['Sales'], s=season_df['Sales'] * 10, c=season_df['Vehicle Type'].map({'Sedan': 'red', 'SUV': 'green', 'Truck': 'blue'}))
plt.title('Impact of Seasonality on Automobile Sales')
plt.xlabel('Season')
plt.ylabel('Sales')
plt.show()
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import pandas as pd
import plotly.graph_objs as go

# Sample data for demonstration
data = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020, 2021],
    'Sedan': [1000, 1200, 1500, 1800, 2000, 1500, 2500],
    'SUV': [800, 1000, 1200, 1400, 1600, 1200, 2000],
    'Truck': [200, 300, 400, 500, 600, 300, 500],
    'Period': ['Non-Recession', 'Non-Recession', 'Non-Recession', 'Non-Recession', 'Non-Recession', 'Recession', 'Recession']
}

df = pd.DataFrame(data)

app = dash.Dash(__name__)

# Layout of the application with dropdowns and output division
app.layout = html.Div([
    html.H1('Automobile Sales Analysis Dashboard', className='title'),
    
    # Dropdown for vehicle type
    dcc.Dropdown(
        id='vehicle-type-dropdown',
        options=[
            {'label': 'Sedan', 'value ': 'Sedan'},
            {'label': 'SUV', 'value': 'SUV'},
            {'label': 'Truck', 'value': 'Truck'}
        ],
        value='Sedan',  # Default value
        clearable=False,
        className='dropdown'
    ),
    
    # Dropdown for period
    dcc.Dropdown(
        id='period-dropdown',
        options=[
            {'label': 'Recession', 'value': 'Recession'},
            {'label': 'Non-Recession', 'value': 'Non-Recession'}
        ],
        value='Recession',  # Default value
        clearable=False,
        className='dropdown'
    ),
    
    # Division for output display
    html.Div(id='output-display', className='output-display'),
    
    # Graph for recession report statistics
    dcc.Graph(id='recession-graph'),
    
    # Graph for yearly report statistics
    dcc.Graph(id='yearly-graph')
])

# Callback to update the recession report graph
@app.callback(
    Output('recession-graph', 'figure'),
    [Input('vehicle-type-dropdown', 'value'),
     Input('period-dropdown', 'value')]
)
def update_recession_graph(selected_vehicle, selected_period):
    # Filter data based on selected period
    filtered_df = df[df['Period'] == selected_period]
    
    # Create the figure
    fig = go.Figure()
    
    # Add the selected vehicle type to the graph
    fig.add_trace(go.Scatter(
        x=filtered_df['Year'],
        y=filtered_df[selected_vehicle],
        mode='lines+markers',
        name=f'{selected_vehicle} Sales'
    ))
    
    # Update the layout
    fig.update_layout(
        title=f'Sales Trend for {selected_vehicle} during {selected_period}',
        xaxis_title='Year',
        yaxis_title='Sales',
        template='plotly'
    )
    
    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import pandas as pd
import plotly.graph_objs as go

# Sample data for demonstration
data = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020, 2021],
    'Sedan': [1000, 1200, 1500, 1800, 2000, 1500, 2500],
    'SUV': [800, 1000, 1200, 1400, 1600, 1200, 2000],
    'Truck': [200, 300, 400, 500, 600, 300, 500],
    'Period': ['Non-Recession', 'Non-Recession', 'Non-Recession', 'Non-Recession', 'Non-Recession', 'Recession', 'Recession']
}

df = pd.DataFrame(data)

# Create a Dash application
app = dash.Dash(__name__)

# Layout of the application with dropdowns and output division
app.layout = html.Div([
    html.H1('Automobile Sales Analysis Dashboard', className='title'),
    
    # Dropdown for vehicle type
    dcc.Dropdown(
        id='vehicle-type-dropdown',
        options=[
            {'label': 'Sedan', 'value': 'Sedan'},
            {'label': 'SUV', 'value': 'SUV'},
            {'label': 'Truck', 'value': 'Truck'}
        ],
        value='Sedan',  # Default value
        clearable=False,
        className='dropdown'
    ),
    
    # Dropdown for period
    dcc.Dropdown(
        id='period-dropdown',
        options=[
            {'label': 'Recession', 'value': 'Recession'},
            {'label': 'Non-Recession', 'value': 'Non-Recession'}
        ],
        value='Recession',  # Default value
        clearable=False,
        className='dropdown'
    ),
    
    # Division for output display
    html.Div(id='output-display', className='output-display'),
    
    # Graph for recession report statistics
    dcc.Graph(id='recession-graph'),
    
    # Graph for yearly report statistics
    dcc.Graph(id='yearly-graph')
])

# Callback to update the recession report graph
@app.callback(
    Output('recession-graph', 'figure'),
    [Input('vehicle-type-dropdown', 'value'),
     Input('period-dropdown', 'value')]
)
def update_recession_graph(selected_vehicle, selected_period):
    # Filter data based on selected period
    filtered_df = df[df['Period'] == selected_period]
    
    # Create the figure
    fig = go.Figure()
    
    # Add the selected vehicle type to the graph
    fig.add_trace(go.Scatter(
        x=filtered_df['Year'],
        y=filtered_df[selected_vehicle],
        mode='lines+markers',
        name=f'{selected_vehicle} Sales'
    ))
    
    # Update the layout
    fig.update_layout(
        title=f'Sales Trend for {selected_vehicle} during {selected_period}',
        xaxis_title='Year',
        yaxis_title='Sales',
        template='plotly'
    )
    
    return fig

# Callback to update the yearly report graph
@app.callback(
    Output('yearly-graph', 'figure'),
    [Input('vehicle-type-dropdown', 'value')]
)
def update_yearly_graph(selected_vehicle):
    # Create the figure for yearly report
    fig = go.Figure()
    
    # Add the selected vehicle type to the graph
    fig.add_trace(go.Bar(
        x=df['Year'],
        y=df[selected_vehicle],
        name=f'{selected_vehicle} Yearly Sales'
    ))
    
    # Update the layout
    fig.update_layout(
        title=f'Yearly Sales for {selected_vehicle}',
        xaxis_title='Year',
        yaxis_title='Sales',
        template='plotly'
    )
    
    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```
